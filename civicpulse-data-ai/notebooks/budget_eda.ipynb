# CivicPulse Budget EDA

This notebook summarizes the cleaned budget dataset, highlights anomaly rows, and prepares a basic handoff for the frontend/backend teams.

```python
import pandas as pd
import json
from pathlib import Path

data_path = Path('../data/processed/clean_budget_data.csv')
df = pd.read_csv(data_path)
df.head()
```

```python
df['alert_fired'] = df['alert_fired'].astype(bool)
df['allocated_ngn'] = pd.to_numeric(df['allocated_ngn'], errors='coerce')
df['actual_ngn'] = pd.to_numeric(df['actual_ngn'], errors='coerce')
df['deviation_pct'] = pd.to_numeric(df['deviation_pct'], errors='coerce')

print('Rows:', len(df))
print('Ministries:', df['ministry'].nunique())
print('Alerts:', int(df['alert_fired'].sum()))
df.groupby('ministry')['allocated_ngn'].agg(['count','mean','max']).head()
```
